```python
import pandas as pd
```

```python
vehicles = pd.read_csv('../datasets/vehicles.csv', sep=',')
```

```python
# Histogram van numerieke 'price' colom
vehicles['price'].plot(kind='hist')
```

```python
vehicles['price'].plot(kind='box')
```

```python
# We hebben te maken met meerdere outliers in deze dataset, dus die moeten er eerst uit gefilterd worden voor deze analyse
# Grotendeels van de 'normale' prijzen liggen onder de 50.000, dus dat wordt de max_price

max_price = 50000
price_filter = vehicles['price'] < max_price
filtered_price = vehicles[price_filter]
```

```python
# Dit plot lijkt op een exponential distribution. Een vrij evenrededige dalende lijn. 

filtered_price['price'].plot(kind='hist')
```

```python
# Histogram van numerieke 'year' colom
vehicles['year'].plot(kind='hist')
```

```python
vehicles['year'].plot(kind='box')
```

```python
# We hebben te maken met meerdere outliers in deze dataset, dus die moeten er eerst uit gefilterd worden voor deze analyse
# Grotendeels van de 'normale' bouwjaren liggen boven de 2.000, dus dat wordt de min_year

min_year = 2000
year_filter = vehicles['year'] > min_year
filtered_year = vehicles[year_filter]
```

```python
# Deze histogram lijkt op een soort omgekeerde lognormal distribution. Hij stijgt vrij geleidelijk en daalt op het einde vrij hard. 

filtered_year['year'].plot(kind='hist')
```
