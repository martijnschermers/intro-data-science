```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder as le
from sklearn import tree
import graphviz

vehicles = pd.read_csv('../datasets/vehicles.csv', sep=',')
```

```python
def calculate_accuracy(predictions, actuals):
    if(len(predictions) != len(actuals)):
        raise Exception("The amount of predictions did not equal the amount of actuals")
    
    return (predictions == actuals).sum() / len(actuals)

def plot_tree_classification(model, features, class_names):
    # Generate plot data
    dot_data = tree.export_graphviz(model, out_file=None, 
                          feature_names=features,  
                          class_names=class_names,  
                          filled=True, rounded=True,  
                          special_characters=True)  

    # Turn into graph using graphviz
    graph = graphviz.Source(dot_data)  

    # Write out a pdf
    graph.render("decision_tree")

    # Display in the notebook
    return graph 
```

```python
# Remove 'nan' values and convert to strings 
vehicles = vehicles[vehicles['condition'] != 'nan']
vehicles['condition'] = vehicles['condition'].astype(str)

vehicles_train, vehicles_test = train_test_split(vehicles, test_size=0.3, random_state=42)
```

```python
vehicles_train = vehicles_train.fillna(vehicles_train['year'].mean())
vehicles_test = vehicles_test.fillna(vehicles_test['year'].mean())
```

```python
features = ['year', 'price']
dt = DecisionTreeClassifier(max_depth = 10) # Increase max_depth to see effect in the plot
dt.fit(vehicles_train[features], vehicles_train['condition'])
```

```python
predictions = dt.predict(vehicles_train[features])
```

```python
# Voorspellingen voor de training set zijn beter dan de test set. 
# Dit verwachte ik wel, omdat het model getrained is op de training set. 
# De test set heeft het model dus nog niet eerder gezien, waardoor de accuracy wat lager was. 

predictionsOnTrainset = dt.predict(vehicles_train[features])
predictionsOnTestset = dt.predict(vehicles_test[features])

accuracyTrain = calculate_accuracy(predictionsOnTrainset, vehicles_train.condition)
accuracyTest = calculate_accuracy(predictionsOnTestset, vehicles_test.condition)

print("Accuracy on training set " + str(accuracyTrain))
print("Accuracy on test set " + str(accuracyTest))
```

```python
plot_tree_classification(dt, features, np.sort(vehicles.condition.unique()))
```
